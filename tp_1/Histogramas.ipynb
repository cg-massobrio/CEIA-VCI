## Histogramas
Con los histogramas puede verse que ambas imágenes poseen la misma distribución de colores. La diferencia está en cómo se presentan en la imagen. Por lo que puede observar de los gráficos, no emplearía "features" a los histogramas para modelos de clasificación.

```python
fig = plt.figure(figsize=(10,8))

ax1=plt.subplot(221)
ax1.imshow(img1_tp, cmap='gray', vmin=0, vmax=255)
ax1.set_title('Imagen 1')

ax2=plt.subplot(222)
ax2.hist(img1_tp.ravel(), bins=40, range=[0, 256], rwidth=1, edgecolor='white')
ax2.set_ylabel('Cantidad de píxeles')
ax2.set_title('Histograma de imagen 1')

ax1=plt.subplot(223)
ax1.imshow(img2_tp, cmap='gray', vmin=0, vmax=255)
ax1.set_title('Imagen 2')

ax2=plt.subplot(224)
ax2.hist(img2_tp.ravel(), bins=40, range=[0, 256], rwidth=1, edgecolor='white')
ax2.set_ylabel('Cantidad de píxeles')
ax2.set_title('Histograma de imagen 2')

fig.tight_layout()
plt.show()
```
